```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dat0=pd.read_csv("5737.spec",names=("wav","flux"),delimiter=" ")
dat1=pd.read_csv("5738.spec",names=("wav","flux"),delimiter=" ")
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(dat0["wav"],dat0["flux"],".")
plt.plot(dat0["wav"],dat0["flux"]*5.e-4,".")
plt.plot(dat0["wav"],dat1["flux"],".")
plt.yscale("log")
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(dat0["wav"],dat1["flux"]/dat0["flux"],".")
plt.axhline(np.median(dat1["flux"]/dat0["flux"]))
plt.yscale("log")
```

```python
plt.hist(np.log10(dat1["flux"]/dat0["flux"]))
```
